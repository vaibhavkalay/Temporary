#### Q-1 LAB DATA AND RENDERED IMAGES 

```python
%matplotlib inline

import scipy as Sci
import scipy.linalg
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
from scipy.signal import hilbert


f = open('LabRFdata.dat','rb')   

Line = np.fromfile(f,count=-1,dtype=np.uint16)
L=len(Line)
f.close() #close the file

x=np.arange(0,L/125e6,1/125e6)  #start,stop,step

plt.subplots_adjust( hspace = 0.5 )
plt.subplot(211)
plt.plot(x,Line[:L])
plt.title('Singe Line Acquired')
plt.xlabel('Time')
plt.ylabel('Unsigned Integer Value')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))


LineF=Line.astype(np.float64)
LineF/=(2**15-1)
LineF-=1.
plt.subplot(414)
plt.plot(x,LineF[:L])
plt.title('Singe Line Acquired in V')
plt.xlabel('Time')
plt.ylabel('V')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
```

#### Q-2 FREQUENCY OF ACQUIRED SIGNAL 

```python
inx1=np.where(x>0.9103e-4)
inx1=inx1[0]
inx2=np.where(x>0.913e-4)
inx2=inx2[0]
plt.plot(x[inx1[0]:inx2[0]],LineF[inx1[0]:inx2[0]])
plt.title('Freequency of the Signals')
plt.xlabel('Time')
plt.ylabel('V')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

# WE GET A FREQUENCY OF 5.5 MHz
```

#### Q-3 BANDWIDTH OF THE SIGNAL 

One definition of bandwidth, for a system, could be the range of frequencies over which the system produces a specified level of performance. A less strict and more practically useful definition will refer to the frequencies beyond which frequency response is small. Small could mean less than 3 dB below the maximum value, or more rarely 10 dB below, or it could mean below a certain absolute value. As with any definition of the width of a function, many definitions are suitable for different purposes.

Bandwidth typically refers to baseband bandwidth in the context of, for example, the sampling theorem and Nyquist sampling rate, while it refers to passband bandwidth in the context of Nyquist symbol rate or Shannon-Hartley channel capacity for communication systems.
